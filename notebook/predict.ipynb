```python
import pandas as pd
import pickle
```

```python
df_clean = pd.read_csv("../data/test.txt") # <== Load you'r DF here
df_target = pd.read_csv("../data/train.txt") # <== Load you'r DF here
```

```python
df_clean= df_clean.dropna()
df_clean.info()

from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()

df_clean['Gender'] = enc.fit_transform(df_clean[['Gender']])
df_clean['Self_Employed'] = enc.fit_transform(df_clean[['Self_Employed']])
df_clean['Education'] = enc.fit_transform(df_clean[['Education']])
df_clean['Dependents'] = enc.fit_transform(df_clean[['Dependents']])
df_clean['Married'] = enc.fit_transform(df_clean[['Married']])
df_OHEncoded = pd.get_dummies(columns = ['Property_Area'],data = df_clean)

df_OHEncoded
```

```python
pipe = pickle.load(open('../DATA/pipe_model_SGB.pkl', 'rb'))
```

```python
from sklearn.metrics import classification_report
import numpy as np

X = df_OHEncoded.drop(['Loan_ID'],axis = 1)
y = df_target['Loan_Status']




y_pred = np.around(pipe.predict(X))

pipe.score(X,y)
# print(pipe.score(X,y))
# print('MAE =',mean_absolute_error(y,y_pred))
# # print('MSE =',mean_squared_error(y,y_pred)/16512)
# print('R2 =',r2_score(y,y_pred))
```
